```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Load pre-trained SentenceTransformer model
model = SentenceTransformer('all-mpnet-base-v2')

# Function to get embeddings from text
def get_embedding(text):
    """
    Converts the input text into a dense vector representation (embedding) using the pre-trained model.

    Args:
    text (str): The input text (description) to be embedded.

    Returns:
    np.array: A vector embedding for the input text.
    """
    return model.encode(text)

# Function to compute cosine similarity between two embeddings
def compute_similarity(embedding1, embedding2):
    """
    Computes the cosine similarity between two embeddings.

    Args:
    embedding1 (np.array): First embedding vector.
    embedding2 (np.array): Second embedding vector.

    Returns:
    float: Cosine similarity score between the two embeddings.
    """
    return cosine_similarity([embedding1], [embedding2])[0][0]

# Function to rank songs based on similarity to the image description
def rank_songs(image_embedding, song_descriptions, top_n=5):
    """
    Ranks the provided songs based on cosine similarity to the image embedding.

    Args:
    image_embedding (np.array): The embedding of the image description.
    song_descriptions (list of str): A list of song descriptions.
    top_n (int): Number of top similar songs to return.

    Returns:
    list of tuple: Top N song descriptions with their similarity scores.
    """
    similarities = []
    for song_description in song_descriptions:
        song_embedding = get_embedding(descriptions)
        similarity = compute_similarity(image_embedding, song_embedding)
        similarities.append((song_description, similarity))

    # Sort songs by similarity in descending order and return the top N
    sorted_songs = sorted(similarities, key=lambda x: x[1], reverse=True)
    return sorted_songs[:top_n]

# Example usage
# if __name__ == "__main__":
#     # Example image and song descriptions
#     image_description = "A beautiful sunset over the mountains, with soft orange and pink hues."
#     song_descriptions = [
#         "A calm acoustic ballad with gentle guitar strumming and soothing vocals.",
#         "A high-energy rock song with fast drums and electric guitars.",
#         "An emotional piano piece with soft melodies and a slow tempo.",
#         "An upbeat pop track with catchy beats and a positive vibe.",
#         "A relaxing instrumental track with nature sounds and ambient tones."
#     ]

#     # Get embeddings for the image description and rank songs
#     image_embedding = get_embedding(image_description)
#     top_n_songs = rank_songs(image_embedding, song_descriptions, top_n=3)

#     # Display the top N songs with similarity scores
#     for idx, (description, similarity) in enumerate(top_n_songs, 1):
#         print(f"Rank {idx}: {description} (Similarity: {similarity:.4f})")
```
